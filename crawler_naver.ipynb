```python
# -*- coding: utf-8 -*-
# author: Jeiyoon
""" 
특이사항: 네이버 검색은 최상의 검색결과를 제공하기 위해, 검색결과를 10페이지까지만 제공하고 있습니다. 
10페이지 내에서 원하는 검색결과를 찾지 못하신 경우, 더욱 구체적인 검색어를 입력해 검색해주세요.
출처: 네이버
"""
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

from urllib.parse import quote

print("[1. 네이버 파워링크 크롤링 코드]")

def crawler(query: str) -> None:
  naver_main_left = 'https://ad.search.naver.com/search.naver?where=ad&query='
  naver_main_right = '&referenceId=hcLpolprvxsssT8y0QsssssstsK-119024'
  link_index = naver_main_left + query + naver_main_right
  print(link_index)

  naver_page = 2 # +1
  naver_start = 1 # +15
  naver_urls = []
  titles = [] # 사이트 이름
  urls = [] # 사이트 주소
  remove_str = '네이버로그인네이버 아이디로 로그인이 가능합니다.서비스 자세히 보기'

  # 총 페이지 수 찾기
  html = urlopen(link_index)
  bsObject = BeautifulSoup(html, 'html.parser')

  idx = bsObject.find_all('div',{'class':'paginate'})

  idx_result = 0
  for i in idx[0]:
    try:
      idx_result += 1
      i.getText()
    except:
      # print("Navigable error")
      idx_result -= 1

  print(idx_result)

  naver_main_left = 'https://ad.search.naver.com/search.naver?where=ad&sm=svc_nrs&query='
  naver_main_right = '&referenceId=hcLpolprvxsssT8y0QsssssstsK-119024&pagingIndex='
  for _ in range(idx_result): 
    link_naver = naver_main_left + query + naver_main_right + str(naver_start)
    naver_urls.append(link_naver)
    naver_start += 1

  print(len(naver_urls))
  # print(naver_urls)
  
  for link in naver_urls:    
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")

    div_s = bsObject.find_all('div', {'class':'ad_section section'})

    for dv in div_s[0].find_all('a', {"class":"lnk_tit"}):
      titles.append(dv.getText())
    for dv in div_s[0].find_all('a', {'class':'url'}):
      urls.append(dv.getText())
 
        
  print(len(urls))    
  print(len(titles))

  for t, u in zip(titles, urls):
    print(t, u)

  
query = quote("감자") # 여기에 검색어를 입력하시면 됩니다 # or 부동산
crawler(query)
```

```python
# -*- coding: utf-8 -*-
# author: Jeiyoon
""" 
특이사항: 네이버 검색은 최상의 검색결과를 제공하기 위해, 검색결과를 10페이지까지만 제공하고 있습니다. 
10페이지 내에서 원하는 검색결과를 찾지 못하신 경우, 더욱 구체적인 검색어를 입력해 검색해주세요.
출처: 네이버
"""
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

from urllib.parse import quote

print("[4. 은행 사이트 크롤링 코드]")

def crawler(query: str) -> None:
  naver_main_left = 'https://search.naver.com/search.naver?display=15&f=&filetype=0&page=2&query='
  naver_main_right = '&research_url=&sm=tab_pge&start=1&where=web'
  link_index = naver_main_left + query + naver_main_right

  naver_page = 2 # +1
  naver_start = 1 # +15
  naver_urls = []
  titles = [] # 사이트 이름
  urls = [] # 사이트 주소
  remove_str = '네이버로그인네이버 아이디로 로그인이 가능합니다.서비스 자세히 보기'

  # 총 페이지 수 찾기
  html = urlopen(link_index)
  bsObject = BeautifulSoup(html, 'html.parser')

  idx = bsObject.find_all('div',{'class':'sc_page_inner'})
  for i in idx:
    idx_result = i.get_text()
    idx_result = list(idx_result.split())
  print(len(idx_result))

  naver_main_left = 'https://search.naver.com/search.naver?display=15&f=&filetype=0&page='
  naver_main_right = '&research_url=&sm=tab_pge&start='
  for _ in range(len(idx_result)-1): # page2 to page 10
    link_naver = naver_main_left + str(naver_page) + '&query=' + query + naver_main_right + str(naver_start) + '&where=web'
    naver_urls.append(link_naver)
    naver_page += 1
    naver_start += 15

  print(len(naver_urls))

  for link in naver_urls:    
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")
    
    li_s = bsObject.find_all('li', {'class':'bx'})

    for li in li_s:
      result = li.find('div',{'class':'total_tit_group'})
      if result is None:
        # print('None Type')
        pass
      else:
        result_link = result.find('a')['href']
        urls.append(result_link)
        result_remove_key = re.sub('\/|:','', result_link)
        result_remove_key = result_remove_key.replace('https','')
        result_remove_key = result_remove_key.replace('http','')
        result_title = result.get_text().replace(remove_str, '')
        result_title = ' '.join(result_title.split())

        if not result_title.startswith('www'): # 링크에 www 가 없는 예외처리
          result_title = 'www.' + result_title

        result_remove_key_except = ''
          
        for s in result_title: # except key 예외처리
          if s != '›':
            result_remove_key_except += s
          else:
            break
          
        if '›' in result_title:
          result_title = result_title.replace(result_remove_key_except, '')
          result_title = result_title.replace('›', '')
          result_title = ' '.join(result_title.split())

          if result_title.startswith('www'): # www. 예외처리
            result_title = re.sub('www. ', '', result_title) 
          titles.append(result_title)
          
        else:
          result_title = result_title.replace(result_remove_key, '')
          result_title = ' '.join(result_title.split())

          if result_title.startswith('www'): # www. 예외처리
            result_title = re.sub('www. ', '', result_title) 
          titles.append(result_title)
        
  print(len(urls))    
  print(len(titles))

  for t, u in zip(titles, urls):
    print(t, u)

  
query = quote("은행 사이트") # 여기에 검색어를 입력하시면 됩니다 # or 부동산
crawler(query)
```
